###### import necessary libraries

```python
import pandas as pd
import numpy as np
df=pd.read_csv('https://raw.githubusercontent.com/amankalal345/DATASETS/main/college_1.csv')
df1=pd.read_csv('https://raw.githubusercontent.com/amankalal345/DATASETS/main/college_2.csv')
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
a=pd.concat([df,df1])
b=a.reset_index(drop=True)
display(b)
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
exceeded_expectations=[]
reached_expectations=[]
needs_improvement=[]
unsatisfactory=[]
for i in b.index:
  if b['CodeKata Score'][i] > 15000:
    exceeded_expectations.append(b.loc[i])
    result1=pd.DataFrame(exceeded_expectations)
    result1.to_csv('Exceeded_expectations.csv', index=False)
  elif 10000 < b['CodeKata Score'][i] < 15000:
    reached_expectations.append(b.loc[i])
    result2=pd.DataFrame(reached_expectations)
    result2.to_csv('Reached_expectations.csv', index=False)
  elif 7000 < b['CodeKata Score'][i] < 10000:
    needs_improvement.append(b.loc[i])
    result3=pd.DataFrame(needs_improvement)
    result3.to_csv('Needs_improvement.csv', index=False)
  else:
    unsatisfactory.append(b.loc[i])
    result4=pd.DataFrame(unsatisfactory)
    result4.to_csv('Unsatisfactory.csv', index=False)
```

```python
d1=pd.read_csv('/content/Exceeded_expectations.csv')
d2=pd.read_csv('/content/Reached_expectations.csv')
d3=pd.read_csv('/content/Needs_improvement.csv')
d4=pd.read_csv('/content/Unsatisfactory.csv')
display(d1)
display(d2)
display(d3)
display(d4)
```
